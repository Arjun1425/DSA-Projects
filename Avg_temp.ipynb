Calculate the average temperature and then find how many days are above average.

```python
def Average_temperature(N_days:int):

    if N_days < 1:
        return f"Your value {N_days} is not correct, please enter the correct value."

    elif N_days == 1:
        temp = float(input("Enter the temperature value: "))
        return f"Your temperature and avg_temp is same: {temp}\u00b0C"

    else:
        temp = [float(input(f"Enter the Day {ele+1} temperature in \u00b0C: ")) for ele in range(N_days)]
        avg_temp = sum(temp) / len(temp)
        high_temp = [1 for ele in temp if ele > avg_temp]
        return f"Average temperature: {avg_temp}, and {sum(high_temp)} Days have more temperature than average."

cal_temp = Average_temperature(5)
print(cal_temp)
```
